```python
import pandas as pd
import numpy as np
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load feature matrix
df_model = pd.read_csv('p2_feature_matrix.csv', parse_dates=['date'])
df_model = df_model.set_index('date').sort_index()

# SHAP validated features only
keep_cols = ['cpi_inflation_lag1', 'cpi_inflation_lag2', 'cpi_inflation_lag3',
             'crude_oil_brent_lag1', 'inr_usd_lag1', 'm3_lag12']

X = df_model[keep_cols]
y = df_model['cpi_inflation']

# Same train test split as all other models
X_train = X[X.index < '2022-01-01']
X_test = X[X.index >= '2022-01-01']
y_train = y[y.index < '2022-01-01']
y_test = y[y.index >= '2022-01-01']

print(X_train.shape, X_test.shape)
```

```python
# Elastic Net has two parameters
# alpha controls overall regularization strength
# l1_ratio controls LASSO vs Ridge balance — 0 is pure Ridge, 1 is pure LASSO
params_en = {
    'alpha': [0.001, 0.01, 0.1, 1.0],
    'l1_ratio': [0.2, 0.5, 0.8, 1.0]
}

tscv = TimeSeriesSplit(n_splits=5)

grid_en = GridSearchCV(ElasticNet(max_iter=10000), params_en,
                       cv=tscv, scoring='neg_root_mean_squared_error',
                       n_jobs=-1)

grid_en.fit(X_train, y_train)

print("Best params:", grid_en.best_params_)
print("Best CV RMSE:", round(-grid_en.best_score_, 4))
```

```python
# Walk forward validation
train_X = X_train.copy()
train_y = y_train.copy()
predictions_en = []

for t in range(len(X_test)):
    # Fit elastic net on expanding window
    model_en = ElasticNet(**grid_en.best_params_, max_iter=10000)
    model_en.fit(train_X, train_y)
    predictions_en.append(model_en.predict(X_test.iloc[[t]])[0])
    train_X = pd.concat([train_X, X_test.iloc[[t]]])
    train_y = pd.concat([train_y, y_test.iloc[[t]]])

rmse_en = np.sqrt(mean_squared_error(y_test, predictions_en))
mae_en = mean_absolute_error(y_test, predictions_en)

print(f"Elastic Net RMSE: {round(rmse_en,4)} MAE: {round(mae_en,4)}")
```

```python
from statsmodels.tsa.arima.model import ARIMA

ext_cols = ['crude_oil_brent_lag1', 'inr_usd_lag1', 'm3_lag12']

# ARIMA residuals
arima_model = ARIMA(y_train, order=(3,0,0)).fit()
residuals = arima_model.resid

# Elastic Net on residuals
en_resid = ElasticNet(**grid_en.best_params_, max_iter=10000)
en_resid.fit(X_train[ext_cols], residuals)

# Walk forward hybrid
train_y = y_train.copy()
predictions_en_hybrid = []

for t in range(len(X_test)):
    arima_pred = ARIMA(train_y, order=(3,0,0)).fit().forecast(steps=1)[0]
    resid_pred = en_resid.predict(X_test[ext_cols].iloc[[t]])[0]
    predictions_en_hybrid.append(arima_pred + resid_pred)
    train_y = pd.concat([train_y, y_test.iloc[[t]]])

rmse_en_hybrid = np.sqrt(mean_squared_error(y_test, predictions_en_hybrid))
mae_en_hybrid = mean_absolute_error(y_test, predictions_en_hybrid)

print(f"ARIMA+ElasticNet Hybrid RMSE: {round(rmse_en_hybrid,4)} MAE: {round(mae_en_hybrid,4)}")
```

```python
pd.DataFrame({
    'elastic_net': predictions_en,
    'en_hybrid': predictions_en_hybrid
}, index=y_test.index).to_csv('p2_en_predictions.csv')
```

```python
# Load all predictions
xgb_pred = pd.read_csv('p2_xgb_predictions.csv', parse_dates=['date'], index_col='date')
en_pred = pd.read_csv('p2_en_predictions.csv', parse_dates=['date'], index_col='date')

# Combine
results_all = pd.concat([xgb_pred, en_pred], axis=1)

fig = px.line(results_all,
              title='Indian CPI Inflation — Model Comparison 2022-2026',
              template='plotly_white',
              height=500)
fig.update_layout(yaxis_title='CPI Inflation %')
fig.show()
```

```python
# Load all predictions
arima_pred = pd.read_csv('p2_arima_predictions.csv', parse_dates=['date'], index_col='date')
xgb_pred = pd.read_csv('p2_xgb_predictions.csv', parse_dates=['date'], index_col='date')
en_pred = pd.read_csv('p2_en_predictions.csv', parse_dates=['date'], index_col='date')

import plotly.graph_objects as go

fig = go.Figure()

# Actual — thick bright white
fig.add_trace(go.Scatter(
    x=xgb_pred.index, y=xgb_pred['actual'],
    name='Actual CPI',
    line=dict(color='#FFFFFF', width=3.5)
))

# ARIMA
fig.add_trace(go.Scatter(
    x=arima_pred.index, y=arima_pred['arima'],
    name='ARIMA(3,0,0)',
    line=dict(color='#00E5FF', width=2.5)
))

# Hybrid — bright green
fig.add_trace(go.Scatter(
    x=xgb_pred.index, y=xgb_pred['hybrid'],
    name='ARIMA+XGBoost Hybrid',
    line=dict(color='#76FF03', width=2)
))

# Elastic Net — orange, thinner
fig.add_trace(go.Scatter(
    x=en_pred.index, y=en_pred['elastic_net'],
    name='Elastic Net',
    line=dict(color='#FF9100', width=1.8)
))

# XGBoost — muted pink, thinnest
fig.add_trace(go.Scatter(
    x=xgb_pred.index, y=xgb_pred['xgboost'],
    name='XGBoost',
    line=dict(color='#FF4081', width=1.2, dash='dot')
))

fig.update_layout(
    title=dict(
        text='Indian CPI Inflation Forecasting — Model Comparison 2022–2026<br>'
             '<sup>Walk-Forward Validation | Test Window January 2022 to 2026</sup>',
        font=dict(size=18, color='white'),
        x=0.05
    ),
    paper_bgcolor='#0D1117',
    plot_bgcolor='#0D1117',
    template='plotly_dark',
    height=550,
    width=1000,
    font=dict(color='white'),
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1,
        font=dict(size=11)
    ),
    yaxis=dict(
        title='CPI Inflation %',
        gridcolor='rgba(255,255,255,0.08)',
        zeroline=False
    ),
    xaxis=dict(
        gridcolor='rgba(255,255,255,0.08)'
    ),
    margin=dict(t=100, b=60, l=60, r=40)
)

# Result annotation
fig.add_annotation(
    x='2023-06-01', y=3.5,
    text='ARIMA RMSE: 0.7027 | Hybrid RMSE: 0.7044 | Elastic Net RMSE: 0.7213 | XGBoost RMSE: 0.8390',
    showarrow=False,
    font=dict(size=10, color='rgba(255,255,255,0.6)'),
    bgcolor='rgba(255,255,255,0.05)',
    bordercolor='rgba(255,255,255,0.1)',
    borderwidth=1,
    borderpad=6
)

fig.show()
```
